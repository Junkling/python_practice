````python
import csv

# ì½”ë”© ì˜ì—­

phonebook = {
    0: {'Name': 'Kim', 'Phone': '45648733'},
    1: {'Name': 'Lee', 'Phone': '89376333'},
    2: {'Name': 'Park', 'Phone': '36457818'},
    3: {'Name': 'Chung', 'Phone': '76891234'},
    4: {'Name': 'Choi', 'Phone': '67451237'}
}

phonebook_list = list(phonebook.values())

def phonebook_print():
    print('=='*20)
    for i in phonebook_list:
        print(f'{i["Name"]} : {i["Phone"]}')
    print('=='*20)

def add_update_member(name:str , phone:str):
    print('=='*20)
    name_list = list(map(lambda v : v.get('Name'),phonebook_list))
    if name_list.__contains__(name):
        print(f'{name}ì˜ ì •ë³´ê°€ ì—…ë°ì´íŠ¸ ë˜ì—ˆìŠµë‹ˆë‹¤.')
        update_phonebook(name , phone)
    else:
        print(f'{name}ê°€ ì¶”ê°€ë˜ì—ˆìŠµë‹ˆë‹¤.')
        phonebook_list.append({'Name': name, 'Phone': phone})
    print('=='*20)

def update_phonebook(name:str , phone_num:str):
    for i in phonebook_list:
        if i.get('Name') == name:
            i.update({'Phone': phone_num})
            break

def get_member_info(name:str)-> str:
    for i in phonebook_list:
        if i.get('Name') == name:
            return f'[{i.get("Name")} : {i.get("Phone")}]'
    else:
        return None

def remove_member(name:str):
    for i in phonebook_list:
        if i.get('Name') == name:
            phonebook_list.remove(i)
            break
    print('=='*20)
    print('ì‚­ì œ ë˜ì—ˆìŠµë‹ˆë‹¤.')
    print('=='*20)

def save_phonebook(type:str):
    print('=='*20)
    if type == 'txt':
        with open('../source/phonebook.txt', 'w') as f:
            for i in phonebook_list:
                f.write(i.get('Name') + ',' + i.get('Phone') + '\n')
        print('ì €ì¥ì´ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤.')
    elif type == 'csv':
        with open('../source/phonebook.csv', 'w') as f:
            csv_col = [['Name' , 'Phone']]
            print(csv_col + list([i.get('Name') , i.get('Phone')] for i in phonebook_list))
            csv.writer(f).writerows(csv_col + list([i.get('Name') , i.get('Phone')] for i in phonebook_list))
        print('ì €ì¥ì´ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤.')
    print('=='*20)

# phonebook_print()

while True:
    choice = input('1: í™•ì¸ , 2: ì¶”ê°€ , 3: ì‚­ì œ , 4: .txt ì €ì¥ 5: .csv ì €ì¥ , 0: ì¢…ë£Œ')
    if choice == '0':
        break

    elif choice == '1':
        phonebook_print()
    elif choice == '2':
        add_name = input('ì¶”ê°€, ìˆ˜ì •í•  ë©¤ë²„ì˜ ì´ë¦„ì„ ì…ë ¥í•˜ì„¸ìš”')
        add_phone = input(f'{add_name}ì˜ ì „í™”ë²ˆí˜¸ë¥¼ ì…ë ¥í•˜ì„¸ìš”')
        add_update_member(add_name , add_phone)
    elif choice == '3':
        remove_name = input('ì‚­ì œí•  ë©¤ë²„ ì´ë¦„ì„ ì…ë ¥í•˜ì„¸ìš”')
        member_info = get_member_info(remove_name)
        if not member_info :
            print('í•´ë‹¹í•˜ëŠ” ìœ ì € ì •ë³´ê°€ ì—†ìŠµë‹ˆë‹¤.')
            continue
        else:
            check_input = input(f'{member_info}ì˜ ì „í™”ë²ˆí˜¸ë¥¼ ì‚­ì œí•˜ì‹œê² ìŠµë‹ˆê¹Œ.(y/n)')
            if check_input == 'y' or 'Y':
                remove_member(remove_name)
            elif check_input == 'n' or 'N':
                continue
    elif choice == '4':
        save_phonebook('txt')
    elif choice == '5':
        save_phonebook('csv')

# print(len(phonebook))
# phonebook.pop(0)
# phonebook.update({0: None})
# print(phonebook)
# print(len(phonebook))
````
