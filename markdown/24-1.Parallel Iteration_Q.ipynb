#### **`ì½”ë”©ì˜ì—­`**

````python
# ì½”ë”© ì˜ì—­
a = ["one", "two", "three", "four"]
b = [30, 20, 15, 75]
c = [5.2, 7.4, 3.6, 4.2]

length = min(len(a), len(b), len(c))

print('='*5 , 'answer1' , '='*5)
answer1 = dict()
for n in range(length):
    # answer1[a[n]] = b[n] * c[n]
    answer1.update({a[n]: b[n] * c[n]})
print(answer1)

print('='*5 , 'answer2' , '='*5)
answer2 = dict()
for item in map(lambda n: {a[n]: b[n] * c[n]}, range(length)):
    # answer1[a[n]] = b[n] * c[n]
    answer2.update(item)
print(answer2)

print('='*5 , 'answer3' , '='*5)
answer3 = {a[n]: b[n] * c[n] for n in range(length)}
print(answer3)

print('='*5 , 'answer4' , '='*5)
answer4 = dict()
# zipì€ ê°™ì€ ì¸ë±ìŠ¤ ê°’ì„ íŠœí”Œë¡œ ë¬¶ìŒ
# print(list(zip(a,b,c)))
for x, y, z in zip(a, b, c):
    answer4[x] = y * z
print(answer4)

print('='*5 , 'answer5' , '='*5)
answer5 = {x: y * z for x, y, z in zip(a, b, c)}
print(answer5)


print('='*5 , 'answer6' , '='*5)
answer6 = dict((x, y * z) for x, y, z in zip(a, b, c))
print(answer6)
````
