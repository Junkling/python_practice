````python
from os import listdir

# ì½”ë”© ì˜ì—­
file_path = "../source/27-1/"

file_list = []
ls = listdir(file_path)
txt_list = list(filter(lambda x: x.endswith('.txt'), ls))

result = []
for file in txt_list:
    with open(file_path + file, 'r') as f:
        result.append(f.read().strip('\n'))
print(result)


import glob

# ê²°ê³¼ ë¦¬ìŠ¤íŠ¸
outputs = []
for file in glob.glob(file_path + '*.txt'):
    with open(file, 'r') as f:
        # ë‚´ìš© í™•ì¸
        # print(f.read())
        outputs.append(f.read().strip('\n')) # .strip('\n')
print(outputs)

# ì‹¤í–‰2
````
